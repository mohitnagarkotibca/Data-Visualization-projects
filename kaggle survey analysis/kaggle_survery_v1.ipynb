```python
#filling with mean and some Bias
def rep(X):
    yrs= df['coding_years'].value_counts().index
    if X==yrs[0]:
        return np.random.choice(np.arange(start=1.1,stop=1.9,step=0.1))
    if X==yrs[1]:
        return np.random.choice(np.arange(start=3.1,stop=5.0,step=0.1))
    if X==yrs[2]:
        return np.random.choice(np.arange(start=5.1,stop=10,step=0.1))
    if X==yrs[3]:
        return np.random.choice(np.arange(start=0.1,stop=0.9,step=0.1))
    if X==yrs[4]:
        return 0
    if X==yrs[5]:
        return np.random.choice(np.arange(start=10.1,stop=20.0,step=0.1))
    if X==yrs[6]:
        return np.random.choice(np.arange(start=20.1,stop=30,step=0.1))
    
df['coding_years']=df['coding_years'].apply(rep)
```

- Their are some people in dataset who have filled 1 options and then submitted the form. So no sense in considering their response.

## Q2: With what education Degree does enthusiast are learning to Explore the data?

ISCED (International Standard Classification of Education) classifies education to mutliple categories
$
\Bigg\{
- Primary education
- Lower secondary education
- Upper secondary education
- Post-secondary non-tertiary education
- Short-cycle tertiary education
\Bigg\}
$
<b>Less than bachelor</b>
<br>
<b>
- Bachelor's or equivalent
- Master's or equivalent
- Doctoral or equivalent 
</b>


Source: https://en.wikipedia.org/wiki/International_Standard_Classification_of_Education

```python
df2= data['Q4'].drop(0)
```

```python
def rep(X):
    if 'Some college/university study without earning a bachelor’s degree' == X:
        return 'less than bachelor'
    elif 'No formal education past high school'== X:
        return 'less than bachelor'
    elif 'I prefer not to answer'==X:
        return 'less than bachelor'
    elif 'Master’s degree'==X:
        return 'Masters'
    elif "Bachelor’s degree"== X:
        return 'Bachelor'
    elif "Doctoral degree"== X:
        return 'Higher than Masters'
    elif "Professional degree"== X:
        return 'Bachelor'

df2= df2.apply(rep).value_counts()
```

```python
df2= df2.loc[['less than bachelor','Bachelor','Masters','Higher than Masters']]
```

```python
plt.figure(figsize=(12,5))
total= np.sum(df2.values)
vals=((df2/total)*100)
plt.barh(df2.index,vals)
plt.xlim((0,100))
plt.ylabel('Education Level')
plt.xlabel('# of Enthusiast')
plt.title('Major Education streams')
plt.annotate(xytext=(62,0.4),xy=(44,2),s='Accounts for \n more than 75% of audience',arrowprops = { 'facecolor' : 'black', 'shrink' : 0.05 })
plt.annotate(xytext=(61,0.86),xy=(44,1),s='',arrowprops = { 'facecolor' : 'black', 'shrink' : 0.05 })
for i,num in enumerate(vals):
    plt.annotate(xy=(np.round(num),i),s=str(round(num))+'%')
    
plt.show()
```
