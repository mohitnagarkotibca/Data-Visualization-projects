```python
#city mileage vs highway mileage fuel efficiency
g = sns.FacetGrid(data = fuel_econ, col = 'VClass', size = 3, col_wrap = 3 )
g = g.map(plt.scatter , 'city' , 'highway', alpha = 0.04) 
# "jitter ="  or "alpha = " are for sb or plt separately 
plt.xlabel('city')
plt.ylabel('highway')
```

Due to overplotting, I've taken a faceting approach to this task. The faceted plots clearly show that most of the high-efficiency cars are in the mid-size and compact car classes.
