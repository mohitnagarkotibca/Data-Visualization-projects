## 4 Data preparation

We have to normalize the variables to express them in the same range of values. In other words, normalization means adjusting values measured on different scales to a common scale.
The points in the normalized data are the same as the original one. The only thing that changes is the scale of the axis.

```python
from sklearn.preprocessing import StandardScaler
ss= StandardScaler()
scaled_data= pd.DataFrame(ss.fit_transform(data.drop('Customer_Segment',axis=1)))
scaled_data.columns= data.drop('Customer_Segment',axis=1).columns
```

```python
scaled_data
```

## 5 k-means execution

In this section we are going to execute the k-means algorithm and analyze the main components that the function returns.

```python
from sklearn.cluster import KMeans
```

```python
#just testing out kmeans cluster
d= scaled_data[['Alcohol','Ash']]
km =KMeans(n_clusters=2).fit(d)
centers= km.cluster_centers_
plt.scatter(x= d['Alcohol'],y= d['Ash'],c= km.labels_)
plt.scatter(x= centers[0][0],y= centers[0][1],marker='X')
plt.scatter(x= centers[1][0],y= centers[1][1],marker='X')
```

```python
km= KMeans(n_clusters=2).fit(scaled_data)
```

```python
scaled_data['target']=km.labels_
```

```python
sns.pairplot(
    data= scaled_data,hue='target',
    markers=['o','s'],
    corner=True
)
```

```python
error=[]
for i in range(1,12):
    km= KMeans(n_clusters=i).fit(scaled_data)
    error.append(km.inertia_)
```

```python
plt.plot(error)
plt.bar(x= np.arange(1,12),height= error,width=0.02)
plt.xticks(ticks=np.arange(1,13))
```
