## Q1: If we take a random person from crowd, what is the probablity that he/she will order a sweet dish?

```python
for_pie= data['flavor_profile'].value_counts()[['spicy','bitter','sweet','sour']]
for_pie= for_pie/np.sum(for_pie)*100

for_pie\
.plot(kind='pie',figsize=(7,7),title='Types of flavour in dataset',shadow=False,explode=(0.1,0.5,0.1,0.2),rotatelabels=True,autopct='%1.1f%%',startangle=290)
plt.ylabel('')
```

Looking at the information in the graph above and applying the concept of theoritical probability, we can make some inferences about the preferred flavors of individuals who participated in the Indian Food 101 experiment. We can conclude that, when selecting an individual from the sample the probability of choosing the Spicy flavor is 58.8% and of choosing Sour is 0.4%.
So, <b> According to dataset, We are 38.9% sure that ,The random individual will choose a sweet dish </b>

Q2: What is the region with most number of dishes ?

```python
data_g= data.groupby('region')
groups= list(data_g.groups.keys())
dish_count=[]
tmp= [dish_count.append(len(data_g.get_group(str(group)))) for group in groups]
plt.figure(figsize=(10,5))
plt.bar(groups,dish_count)

for i,group in enumerate(groups):
    plt.annotate(s=str(dish_count[i]),xy=(str(group),dish_count[i]+1))
plt.title('Indian Regions with most number of dishes')
```

From the above figure it can be seen that West region consisting Gujarat, Rajasthan, Punjab has the most number of dishes in the dataset, Where central region consisting of madhya pradesh, upper Maharashtra has lowest number of dishes

Q3: What is the statewise distribution of indian dishes?

```python
plt.figure(figsize=(12,7))
data_s= data.groupby('state')
states= list(data_s.groups.keys())
state_dish=[]
tmp= [state_dish.append(len(data_s.get_group(str(state)))) for state in states]
state_dish= np.round(state_dish/np.sum(state_dish)*100,2)
plt.barh(states,state_dish)
for i,state in enumerate(states):
    plt.annotate(s=str(state_dish[i])+'%',xy=(state_dish[i],state))
plt.title('State wise distribution of dishes')
```

Q4: What are the commonly used ingredients region wise ?

```python
data_rg= data.groupby('region')
for_count=[]
word_clouds=[]
regions= list(data_rg.groups.keys())
for region in regions:
    data_rg= data.groupby('region')
    ingredients= list(data_rg.get_group(str(region))['ingredients'])
    d= [[for_count.append(i.strip()) for i in ingredients[j].split(',')] for j in range(len(ingredients))]
    counts= pd.Series(for_count).value_counts()
    wc= WordCloud(background_color='#e8eec3',stopwords=STOPWORDS)
    zz=word_clouds.append(wc.generate(' '.join(for_count)))
fig,ax= plt.subplots(nrows=4,ncols=2,figsize=(20,20))
l=0
for i in range(2):
    for j in range(4):
        if l==7:
            break
        ax[j,i].imshow(word_clouds[l])
        ax[j,i].set_title('For '+str(regions[l])+' region')
        l+=1
```

Q5: Most common ingredients in vegetarian food 

```python
fig,ax= plt.subplots(nrows=1,ncols=2,figsize=(20,8))

data_diet_v= data.groupby('diet').get_group('vegetarian')
data_diet_n= data.groupby('diet').get_group('non vegetarian')

veg_ingredients=[]
nonveg_ingredients=[]
tmp= [[veg_ingredients.append(i.strip()) for i in item.split(',')] for item in data_diet_v['ingredients']]
tmp= [[nonveg_ingredients.append(i.strip()) for i in item.split(',')] for item in data_diet_n['ingredients']]

ax[0].imshow(wc.generate(' '.join(veg_ingredients)))
ax[0].set_title('Common ingredients in vegetarian foods')

ax[1].imshow(wc.generate(' '.join(nonveg_ingredients)))
ax[1].set_title('Common ingredients in non vegetarian foods')
```

```python
mah_data = data[data['state']=='Maharashtra']
total_dishes = len(mah_data)

course_data = mah_data['course'].value_counts().reset_index()
course_data.columns = ['course','num']

diet_data = mah_data['diet'].value_counts().reset_index()
diet_data.columns = ['diet','num']


fig= make_subplots(rows=2,cols=2,subplot_titles=('total dishes','main courses according to dishes','Vegetarian and Non vegetarian proportions',''),
                  specs=[
                      [
                          {'type':'indicator'},
                          {'type':'bar','rowspan':2}
                      ],
                          [
                              {'type':'pie'},
                              {'type':'pie'}
                          ]
                  ])

fig.add_trace(go.Indicator(
    mode = 'number',
    value = int(total_dishes),
    number={'font':{'color': 'blue','size':100}},
),row=1, col=1)



fig.add_trace(go.Bar(
    x=course_data['course'],y=course_data['num'],
    marker={'color': 'orange'},
    text=course_data['num'],
    name='dishes by courses',
    textposition ='auto',
    showlegend=False),row=1, col=2)

fig.add_trace(go.Pie(labels=diet_data['diet'], 
                     values=diet_data['num'],textinfo='percent',
                     marker= dict(colors=['green','red']),
                     showlegend=True
                    )
              ,row=2, col=1)

fig.update_layout(
    title_text='Maharashtra Food-Mini Infograph',template='plotly',
    title_x=0.5,
    showlegend=True
)

fig.show()
```

```python
data_wb= data[data['state']=='West Bengal']
counts= len(data_wb)
v_c= data['course'].value_counts().reset_index()
veg_non_veg= data['diet'].value_counts()

fig= make_subplots(rows=2,
                   cols=2,
                   subplot_titles=['Total dishes from West Bengal','Variety Available','Veg and Non Veg proportions',''],
                   specs=[
                       [
                       {'type':'indicator'},
                       {'type':'bar','rowspan':2}
                       ],
                       [
                        {'type':'pie'},
                        {'type':'pie'}
                       ]
                   ]
                  )
fig.add_trace(go.Indicator(mode='number',value=counts,number={'font':{'color': 'blue','size':100}}),row=1,col=1)
fig.add_trace(go.Bar(x= v_c['index'],y=v_c['course'],marker={'color':'orange'},showlegend=False),row=1,col=2)
fig.add_trace(go.Pie(labels=list(veg_non_veg.index),values=veg_non_veg.values,marker= dict(colors=['red','green']),textinfo='percent',rotation=70),row=2,col=1)
fig.update_layout(title='West bengal Mini food graph',title_x=0.5,template='plotly',showlegend=True)

fig.show()
```

```python
data_pu= data[data['state']=='Punjab']
counts= len(data_pu)
pie_pu= data_pu['diet'].value_counts()
bar_pu= data_pu['course'].value_counts()
fig= make_subplots(rows=2,
                   cols=2,
                   subplot_titles=['Toal Dishes from Punjab','Variety Available','Veg and non veg proportions',''],
                   specs=[[{'type':'indicator'},
                         {'type':'bar','rowspan':2}],
                         [{'type':'pie'},{'type':'pie'}]
                         ]
                  )
fig.add_trace(go.Indicator(mode='number',value=counts,number={'font':{'color':'blue','size':100}}),row=1,col=1)
fig.add_trace(go.Bar(x= bar_pu.index,y=bar_pu.values,marker={'color':'orange'},showlegend=True),row=1,col=2)
fig.add_trace(go.Pie(labels=pie_pu.index,values=pie_pu.values,textinfo='percent',marker=dict(colors=['red','green']),rotation=90),row=2,col=1)
fig.update_layout(title='Punjab Mini food graph',title_x=0.5,template='plotly',showlegend=True)
fig.show()
```
